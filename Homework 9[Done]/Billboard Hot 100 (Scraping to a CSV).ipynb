```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

```python
items = doc.find_all('div', class_='chart-list-item')
```

```python
for item in items:
    print(item['data-title'], item['data-artist'], item['data-rank'])
```

```python
rows = []

for item in items:
    row = {}
    row['Rank'] = item['data-rank']
    row['Title'] = item['data-title']
    row['Artist'] = item['data-artist']
    
    rows.append(row)

df = pd.DataFrame(rows)
df
   
```

```python
df.to_csv("billboard.csv", index=False)
```
