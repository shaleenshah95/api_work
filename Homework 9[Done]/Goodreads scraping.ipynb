```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
ranks = doc.find_all('td', class_ = 'number')

for rank in ranks:
    print(rank.text)
```

```python
titles = doc.find_all('a', class_ = 'bookTitle')

for title in titles:
    print(title.find_all('span')[0].text)
    
```

```python
authors = doc.find_all('a', class_='authorName')

for author in authors:
    print(author.text)
```

```python
scores = doc.find_all('span', class_='smallText uitext')

for score in scores:
    print(score.text)
```
