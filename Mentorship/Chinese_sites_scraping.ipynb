```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url0 = "https://www.nychinaren.com/company/catid_33/%E6%9C%88%E5%AD%90%E4%B8%AD%E5%BF%83/task_list.html"
url1 = "https://www.nychinaren.com/company/catid_33/%E6%9C%88%E5%AD%90%E4%B8%AD%E5%BF%83/task_list/page_2.html"
url2 = "https://www.nychinaren.com/company/catid_33/%E6%9C%88%E5%AD%90%E4%B8%AD%E5%BF%83/task_list/page_3.html"
url3 = "https://www.nychinaren.com/company/catid_33/%E6%9C%88%E5%AD%90%E4%B8%AD%E5%BF%83/task_list/page_4.html"
url4 = "https://www.nychinaren.com/company/catid_33/%E6%9C%88%E5%AD%90%E4%B8%AD%E5%BF%83/task_list/page_5.html"

html0 = requests.get(url0).content
html1 = requests.get(url1).content
html2 = requests.get(url2).content
html3 = requests.get(url3).content
html4 = requests.get(url4).content

doc0 = BeautifulSoup(html0, "html.parser")
doc1 = BeautifulSoup(html1, "html.parser")
doc2 = BeautifulSoup(html2, "html.parser")
doc3 = BeautifulSoup(html3, "html.parser")
doc4 = BeautifulSoup(html4, "html.parser")
```

```python
items0 = doc0.find_all('dl', class_='regular_company')
items1 = doc1.find_all('dl', class_='regular_company')
items2 = doc2.find_all('dl', class_='regular_company')
items3 = doc3.find_all('dl', class_='regular_company')
items4 = doc4.find_all('dl', class_='regular_company')

companies0 = []
companies1 = []
companies2 = []
companies3 = []
companies4 = []

for item0 in items0:
    try:
        companies0.append(item0.find('a', class_='map_link')['title'])
    except:
        pass
    
for item1 in items1:
    try:
        companies1.append(item1.find('a', class_='map_link')['title'])
    except:
        pass
    
for item2 in items2:
    try:
        companies2.append(item2.find('a', class_='map_link')['title'])
    except:
        pass
    
for item3 in items3:
    try:
        companies3.append(item3.find('a', class_='map_link')['title'])
    except:
        pass
    
for item4 in items4:
    try:
        companies4.append(item4.find('a', class_='map_link')['title'])
    except:
        pass
        
companies = companies0 + companies1 + companies2 + companies3 + companies4

companies
```

```python
addresses0=[]
addresses1=[]
addresses2=[]
addresses3=[]
addresses4=[]

for item0 in items0:
    address_html0 = item0.find('div', class_='adress_text')
    try:
        addresses0.append(address_html0.text)
    except:
        addresses0.append('No address.')

for item1 in items1:
    address_html1 = item1.find('div', class_='adress_text')
    try:
        addresses1.append(address_html1.text)
    except:
        addresses1.append('No address.')

for item2 in items2:
    address_html2 = item2.find('div', class_='adress_text')
    try:
        addresses2.append(address_html2.text)
    except:
        addresses2.append('No address.')

for item3 in items3:
    address_html3 = item3.find('div', class_='adress_text')
    try:
        addresses3.append(address_html3.text)
    except:
        addresses3.append('No address.')
        
for item4 in items4:
    address_html4 = item4.find('div', class_='adress_text')
    try:
        addresses4.append(address_html4.text)
    except:
        addresses4.append('No address.')
        
addresses = addresses0 +addresses1 +addresses2 +addresses3 +addresses4

addresses
```
